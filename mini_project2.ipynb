```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/Vehicle_Insurance.csv")
```

```python
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Response"].fillna(df["Response"].mode()[0],inplace=True)
```

```python
df.columns
```

```python
sns.countplot(x='Response', data=df)
plt.title("Distribution of Insurance Claims")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Annual_Premium'], bins=30, kde=True)
plt.title("Annual Premium Distribution")
plt.xlabel("Premium Amount")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Vintage'], bins=30, kde=True)
plt.title("Customer Vintage Distribution")
plt.xlabel("Days")
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='Vehicle_Age', data=df)
plt.title("Vehicle Age Distribution")
plt.show()
```

```python
sns.countplot(x='Vehicle_Damage', data=df)
plt.title("Vehicle Damage Distribution")
plt.show()
```
